<b>Data Cleaning

```python
df[['Genre','Hidden Gem Score','Runtime','IMDb Score','Rotten Tomatoes Score','Metacritic Score',]].isnull().sum()
```

```python
df['Genre'] = df['Genre'].fillna(df['Genre'].mode()[0])
```

```python
df['Hidden Gem Score'] = df['Hidden Gem Score'].fillna(df['Hidden Gem Score'].median())
df['IMDb Score'] = df['IMDb Score'].fillna(df['IMDb Score'].median())
df['Rotten Tomatoes Score'] = df['Rotten Tomatoes Score'].fillna(df['Rotten Tomatoes Score'].median())
df['Metacritic Score'] = df['Metacritic Score'].fillna(df['Metacritic Score'].median())
df['Runtime'] = df['Runtime'].fillna(0)
```

```python
df[['Genre','Hidden Gem Score','Runtime','IMDb Score','Rotten Tomatoes Score','Metacritic Score',]].isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

<b>Data Visualization

```python
director_group = df.groupby('Director')['IMDb Score'].mean().reset_index()
director_group = director_group.sort_values(by='IMDb Score', ascending=False).head(5)

print(director_group)
plt.figure(figsize=(10,5))
plt.bar(director_group['Director'], director_group['IMDb Score'], color='skyblue', edgecolor='black')

# Adding labels and title
plt.title('Top 5 Directors by Average IMDb Score', fontsize=14, fontweight='bold')
plt.xlabel('Director', fontsize=12)
plt.ylabel('Average IMDb Score', fontsize=12)
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

*Interpretation*
* Dorin Marcu achieved the highest average IMDb score (9.7), indicating exceptional audience ratings.

*  All top 5 directors have scores above 9.0, showing consistently high-quality direction.

*  The difference in scores among them is minimal, suggesting strong competition in performance.

*  Collaborative directors (e.g., Bill Butler & William A. Graham) also maintain high viewer ratings.

```python
sns.histplot(df['IMDb Score'], kde=True)
plt.title("Distribution of IMDB Scores")
```

*Interpretation*

*  Very few movies score below 4 or above 8, showing that extreme ratings are rare.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Boxoffice', 
    y='IMDb Score', 
    data=df, 
    color='teal', 
    edgecolor='black', 
    alpha=0.6
)

plt.title('Box Office vs IMDb Score', fontsize=14, fontweight='bold')
plt.xlabel('Box Office (in millions)')
plt.ylabel('IMDb Score')


plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
```

*Interpretation*

*  Moderate links: IMDb aligns moderately with critic scores; Awards relate modestly to Metacritic.

*  Weak links: Hidden Gem Score shows little correlation with others, slightly negative with Awards.

```python
df['Netflix Release Date'] = pd.to_datetime(df['Netflix Release Date'], errors='coerce')
df['Netflix Release Date'] = df['Netflix Release Date'].dt.year
```

```python
# Group by year and take mean IMDb Score
yearly_trend = df.groupby('Netflix Release Date')['IMDb Score'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(
    x='Netflix Release Date', 
    y='IMDb Score', 
    data=yearly_trend, 
    marker='o', 
    color='teal'
)

plt.title('Trend of Average IMDb Score Over Years', fontsize=14, fontweight='bold')
plt.xlabel('Release Year')
plt.ylabel('Average IMDb Score')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

*Interpretation*
*  Overall trend: Average IMDb scores declined from 2015 to 2020, reaching the lowest point in 2020.

*  Slight recovery: A small uptick is seen in 2021, suggesting mild improvement.

*  Insight: Viewer ratings have generally decreased over time, possibly indicating lower perceived quality or changing audience standards.

<b>Outliers Detection

```python
sns.boxplot(df['IMDb Score'])
plt.title("Outlier Detection - IMDB Score")
```

*Interpretation*
*  Central tendency: Most IMDb scores cluster around 6–7.

*  Spread: The middle 50% of scores lie roughly between 6 and 7.5.

*  Outliers: A few movies score below 4 or above 9, indicating rare extremes in audience ratings.

*  IMDb scores are roughly symmetric but slightly left-skewed, indicating a small tendency toward lower ratings.

*  Insight: Ratings are fairly consistent, with only a small portion of exceptionally poor or excellent films.

```python
Q1 = df['IMDb Score'].quantile(0.25)
Q3 = df['IMDb Score'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['IMDb Score'] >= Q1 - 1.5*IQR) & (df['IMDb Score'] <= Q3 + 1.5*IQR)]
```

```python
sns.boxplot(df['IMDb Score'])
plt.title("After Outlier Detection - IMDB Score")
```

*Interpretation*
*  Outliers removed: Extreme low and high IMDb scores are eliminated.

*  Central range: Most scores now lie between 6.0 and 7.5.

*  Distribution: The data looks more balanced and symmetric, with reduced skewness.

*  Insight: IMDb ratings are now tightly centered around the average, giving a clearer picture of typical movie performance.
